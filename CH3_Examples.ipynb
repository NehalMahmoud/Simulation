## Algebric properties

## 1- Addition of matrices

#### ex1

```julia
A=[1 2 3;
   0 4 2]

B=[2  3  4;
   0 -1 -4] 

C=A+B
```

## 2- Scalar Multiplication

#### ex2

```julia
A=[2 1  0;
   1 4 -1]

B=2*A
```

## 3- Matrix Multiplication 

#### ex3

```julia
A=[1 -2 0;
   2  1 3]

B=[1 -3  3 0;
   1  4 -1 0;
   2  1  2 1]

C=A*B
```

## 4- Transpose

```julia
A=[1 2 3;
   0 4 5]
A'
```

## 5- Determinant

```julia
A=[3 2 1 0;
   1 0 0 2;
   1 0 2 1;
   4 2 5 0]

det(A)
```

```julia
A=[1 0  0  0 
   5 2  0  0
   6 7  3  0 
   9 10 11 4]
det(A)
```

## 6- Matrix Inverse

```julia
A=[1 0 2
   3 1 0
   0 1 4]

inv(A)
```
